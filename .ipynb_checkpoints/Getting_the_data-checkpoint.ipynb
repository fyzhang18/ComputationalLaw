```python
# Here we will put all the packages needed. All lines that starts with an exclamation mark (!) are packages to download (not 
# very common). If you don't have the package, just remove the comment symbol (#)

#! pip install python-docx

import os

from docx import Document

path_to_folder = '/Users/Felicia/Dropbox/Predicting Settlement Value/'
os.chdir('{}154 cases'.format(path_to_folder))
```

```python
# list document names

os.listdir()
```

```python
# This code creates a dictionary called documents, where the files are stored. 
# Each document has a number as its index. 

# MAKE SURE NO DOCUMENTS ARE OPEN or else will get error.

# Initializing the dictionary
documents = {}
# Setting initial value for the first document
i = 0

# Initializing for loop over all files in folder
for filename in os.listdir():
    # Initialize document
    doc = ''
    # Create a temporary file. Within this file we will get every paragraph 
    temp = Document('{}'.format(filename))
    
    # Create a for loop for every paragraph in the file to form the complete file
    for p in temp.paragraphs:
        
        doc = doc + ' ' + p.text

    documents[i] = doc
    
    i+=1
```

```python
#! pip install nltk

# We are going to use the NLTK package for Natural language Processing. 

# For the first model we will use the word tokenizer. 
# This means that we are going to treat words separately. The other choice would be sentence tokenizer. 
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation

nltk.download('punkt')
nltk.download("stopwords")

# This is our own tokenizer
from nltk.tokenize import RegexpTokenizer
```

```python
# Variables

# This tokenizer eliminates. 
# This grabs every single word and splitting it and ignoring punctuations
tokenizer = RegexpTokenizer(r'\w+')

# https://regex101.com/r/cO8lqs/4
```

```python
# Clean the file!

# use a different tokenizer, but keep the stop words

def clean_file(file):
    # Grabs single words
    clean_file = tokenizer.tokenize(file)
    # Check if word is a stop word
    clean_file = [x for x in clean_file if x not in stopwords.words()]
    
    final_file = []

    for item in clean_file:
        # This checks if the word has more than one letter
        if len(item) != 1:
            # This code eliminates numbers that are in string format...not good for our case
            try:
                int(item)

            except:
                final_file.append(item)

    return final_file
```

```python
stopwords.words()
```

```python
# Now let's tokenize each document. this takes a while....
tokenized_documents = {}
i = 0
for item in documents.keys():
    print(i)
    
    i += 1
    
    tokenized_documents[item] = clean_file(documents[item])

```

```python
# Compare this tokenizer with our custom one.
print(word_tokenize(documents[0]))
print('#######################')
print('#######################')
print('#######################')
print('#######################')
print(' ')
print(' ')
print(clean_file(documents[0]))
```

```python
# Now let's construct the vocabulary used in all documents (AKA getting every single unique word in every document) 
# We loop through all tokenized documents and add each words. Then we eliminate the duplicates, constructing the vocab. 
# This will be the columns of our TDI-DF matrix (That matrix that I've shown you with each document as each row).
vocab = []
for item in tokenized_documents.keys():
    vocab = vocab + tokenized_documents[item]

vocab = set(vocab)
print(len(vocab))
```
